```python
import os
import sys
import yaml
import requests
import pandas as pd
from sdk import Sedmax, ElectricalArchive

module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

config_path = os.path.join(module_path, './config/config.yaml')
```

```python
SEDMAX_config = yaml.safe_load(open(config_path))['vars']

s = Sedmax(SEDMAX_config['SEDMAX_URL'])
s.login(SEDMAX_config['SEDMAX_USERNAME'], SEDMAX_config['SEDMAX_PASSWORD'])
el = ElectricalArchive(s)
```

```python
df = el.get_data([SEDMAX_config['dev-101']], ['30min'], '2022-03-01', '2022-03-21').reset_index()
```

```python
#
r = requests.get('https://isdayoff.ru/api/getdata?year=2022')
days_off = [x for x in r.text]
days_off = dict(enumerate(days_off))
```

```python
df['dayofweek'] = df['dt'].dt.dayofweek.astype(str)
df['dayoff'] = df['dt'].dt.day_of_year.map(days_off)
df['hour'] = df['dt'].dt.hour.astype(str)
pd.get_dummies(df)
```

```python
import plotly.graph_objects as go

fig = go.Figure(
    data=[go.Scatter(x = df.dt, y=df['dev-101_ea_imp'])],
    layout=dict(title=dict(text="A Figure Specified By A Graph Object"))
)

fig.show()
```
